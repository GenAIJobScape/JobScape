## 시각화 1: AI 업무 활용자(IT/개발자)의 위협 인식과 구직자 불안감
### 내용:
- 전반적인 한국인 인식과 AI를 실제로 다루는 직군(IT/개발자)의 높은 위협 인식을 대조하고, 이것이 구직자들의 행동 변화로 이어진다는 흐름을 보여줍니다.

### 목적:
AI 위협 인식이 단순한 막연한 불안감이 아니라, AI를 직접 다루는 직군에서 더 높게 체감되며, 이것이 구직자들의 취업 전략 수정이라는 구체적인 행동 변화로 이어진다는 것을 강조합니다.

```python
# 시각화 1: AI 업무 활용자(IT/개발자)의 위협 인식과 구직자 불안감

# 데이터 준비
df_ai_threat_perception = pd.DataFrame({
    'Category': ['한국인 AI 업무 위협 인식', 'AI 업무 활용자 역할 대체 가능성 인식', '구직자 취업 전략 수정 필요성'],
    'Value': [korean_ai_threat_perception,
              df_personal_role_replace_user[df_personal_role_replace_user['Category'] == '대체할 수 있음']['Value'].iloc[0],
              job_seeker_strategy_change],
    'Group': ['인식', '인식', '행동 변화']
})

plt.figure(figsize=(10, 7))
sns.barplot(x='Category', y='Value', data=df_ai_threat_perception, hue='Group', palette=['skyblue', 'lightcoral', 'lightgreen'], dodge=False) # dodge=False로 막대 겹치지 않게
plt.title('AI 위협 인식과 구직자 취업 전략 변화의 연관성', fontsize=16)
plt.xlabel('')
plt.ylabel('비율', fontsize=12)
plt.gca().yaxis.set_major_formatter(mticker.PercentFormatter(xmax=1.0))
plt.ylim(0, 0.8)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for index, row in df_ai_threat_perception.iterrows():
    plt.text(row.name, row.Value + 0.02, f'{row.Value:.1%}', color='black', ha="center", va='bottom', fontsize=10)

plt.legend().set_visible(False) # 범례 숨기기
plt.xticks(wrap=True) # 긴 라벨 처리
plt.tight_layout()
plt.show()

print("▶ 시사점: 한국인 전반의 AI 위협 인식(35.4%)을 넘어, AI를 실제로 활용하는 IT/개발자 직군에서는 본인 역할 대체 가능성 인식(48.8%)이 훨씬 높게 나타났습니다. 이러한 위협 인식은 구직자의 73%가 취업 전략 수정을 필요하다고 느끼는 구체적인 행동 변화로 이어지고 있음을 보여줍니다.")
```

## 시각화 2: IT/개발 직군 내 AI 대체 위협 체감 (업무 활용자, 관리직) 및 예상 시기
### 내용:
- IT/개발자 직군의 실무자와 관리자 모두 AI 대체 위협을 체감하고 있음을 보여주며, 특히 구성원 업무의 대체가 더 빠를 것이라는 인식을 통해 기업의 채용 보류 경향을 간접적으로 뒷받침합니다.

###목적:
- AI 위협 인식이 IT/개발 직군 내에서도 보편적이며, 단기적/중기적인 변화가 예상됨을 강조합니다.

```python
# 시각화 2: IT/개발 직군 내 AI 대체 위협 체감 및 예상 시기

# 데이터 준비
df_role_replace_comparison = pd.DataFrame({
    '그룹': ['AI 업무 활용자 (본인)', 'AI 업무 활용 관리직 (구성원)'],
    '대체할 수 있음': [df_personal_role_replace_user[df_personal_role_replace_user['Category'] == '대체할 수 있음']['Value'].iloc[0],
                   df_subordinate_role_replace_user_manager[df_subordinate_role_replace_user_manager['Category'] == '대체할 수 있음']['Value'].iloc[0]],
    '5년 이내 대체 예상': [df_personal_replace_time_user[df_personal_replace_time_user['Period'] == '5년 이내']['Value'].iloc[0],
                        df_subordinate_replace_time_user_manager[df_subordinate_replace_time_user_manager['Period'] == '5년 이내']['Value'].iloc[0]],
    '10년 이내 대체 예상': [df_personal_replace_time_user[df_personal_replace_time_user['Period'].isin(['5년 이내', '6년-10년 이내'])]['Value'].sum(),
                         df_subordinate_replace_time_user_manager[df_subordinate_replace_time_user_manager['Period'].isin(['5년 이내', '6년-10년 이내'])]['Value'].sum()]
})

df_plot_melted = df_role_replace_comparison.melt(id_vars='그룹', var_name='지표', value_name='비율')

plt.figure(figsize=(12, 7))
sns.barplot(x='그룹', y='비율', hue='지표', data=df_plot_melted, palette='coolwarm')
plt.title('AI 업무 활용 직군 (IT/개발자)의 업무 대체 위협 체감', fontsize=16)
plt.xlabel('')
plt.ylabel('비율', fontsize=12)
plt.gca().yaxis.set_major_formatter(mticker.PercentFormatter(xmax=1.0))
plt.ylim(0, 1.0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for p in plt.gca().patches:
    if p.get_height() > 0:
        plt.gca().annotate(f'{p.get_height():.1%}',
                           (p.get_x() + p.get_width() / 2., p.get_height()),
                           ha='center', va='center', xytext=(0, 5), textcoords='offset points', fontsize=9)

plt.legend(title='인식/예상', loc='upper right', bbox_to_anchor=(1.25, 1))
plt.tight_layout()
plt.show()

print("▶ 시사점: AI를 활용하는 IT/개발 실무자(48.8%)와 관리자(47.9%) 모두 AI가 자신의 업무를 대체할 수 있다고 유사하게 인식합니다. 특히, 관리직은 구성원 업무가 10년 이내에 대체될 가능성(82.3%)을 높게 보고 있어, 기업 내부에서 AI로 인한 인력 구조 개편이 활발하게 논의될 수 있음을 보여줍니다.")
```

## 시각화 3: AI 대체 위협 인식과 실제 IT 직무 채용 감소의 상관 관계

### 내용:
- 높은 AI 대체 위협 인식이 특정 IT 직무의 실제 일자리 감소로 이어지고 있음을 명확히 보여줍니다.
